```python
import astropy.units as u
from astropy.coordinates import SkyCoord
from astroquery.gaia import Gaia
import matplotlib.pyplot as plt
```

```python
coord = SkyCoord(ra=280, dec=-60, unit=(u.degree, u.degree))
width = u.Quantity(0.1, u.deg)
height = u.Quantity(0.1, u.deg)
```

```python
r = Gaia.query_object_async(coordinate=coord, width=width, height=height)
```

```python
ra = r['ra']
dec = r['dec']
```

```python
ra = r['ra']
dec = r['dec']

plt.figure(figsize=(8, 6))
plt.scatter(ra, dec, s=1, color='blue')
plt.title('Star Positions in RA/Dec')
plt.xlabel('Right Ascension (degrees)')
plt.ylabel('Declination (degrees)')
plt.grid(True)
plt.show()
```

```python
print(len(r), len(r.colnames))
print(r.colnames)
```

```python
Gaia.ROW_LIMIT = 100

coord = SkyCoord(ra=280, dec=-60, unit=(u.degree, u.degree))
width = u.Quantity(1, u.deg)
height = u.Quantity(1, u.deg)
r2 = Gaia.query_object_async(coordinate=coord, width=width, height=height)
print(len(r2), len(r2.colnames))
```

```python
ra = r2['ra']
dec = r2['dec']

plt.figure(figsize=(8, 6))
plt.scatter(ra, dec, s=1, color='blue')
plt.title('Star Positions in RA/Dec')
plt.xlabel('Right Ascension (degrees)')
plt.ylabel('Declination (degrees)')
plt.grid(True)
plt.show()
```
